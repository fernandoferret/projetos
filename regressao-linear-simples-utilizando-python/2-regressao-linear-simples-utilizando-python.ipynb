```python
# Importar bibliotecas
# Para criar o modelo de regressão: sklearn
# Para visualização de residuais: yellowbrick
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot 
```

```python
# Carregar dados
base = pd.read_csv('cars.csv')
base.shape
```

```python
# Visualizar dados
base.head()
```

```python
# Excluir primeira coluna
base = base.drop(['Unnamed: 0'], axis = 1)
base.head()
```

```python
# Definir variáveis X e Y
# X - distância - variável independente
# Y - velocidade - variável dependente
# A partir da distância de frenagem obter a velocidade no momento da frenagem
X = base.iloc[:, 1].values
y = base.iloc[:, 0].values
```

```python
# Visualizar X
X
```

```python
# Calcular correlação entre X e Y
correlacao = np.corrcoef(X, y)
correlacao
```

```python
# Formato de matriz com uma coluna a mais
X = X.reshape(-1, 1)
X
```

```python
# Criação do modelo e treinamento
# Fit indica que o treinamento deve ser executado
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Visualizar coeficientes
modelo.intercept_
```

```python
# Visualizar inclinação
modelo.coef_
```

```python
# Geração do gráfico com os pontos reais e as previsões
plt.scatter(X, y)
plt.plot(X, modelo.predict(X), color = 'green')
```

```python
# Previsão da "distância 22 pés" usando a fórmula manual
# interceptação + inclinação * valor de 'dist'
# A qual velocidade estava se percorreu 22 pés até parar?
modelo.intercept_ + modelo.coef_ * 22
```

```python
# Previsão utilizando função do sklearn
modelo.predict([[22]])
```

```python
# Visualizar gráfico com os residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(X, y)
visualizador.poof()
```
